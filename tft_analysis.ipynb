```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.sans-serif'] = ['SimHei']  
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_csv('./TFT_Challenger_MatchData.csv')  

print("数据有几行几列：", df.shape)
print("\n前5行数据预览：")
print(df.head())
print("\n基本统计：")
print(df.describe())
```

```python
import matplotlib.pyplot as plt

df['win'] = (df['Ranked'] == 1).astype(int)  # 1 = 吃鸡

win_by_level = df.groupby('level')['win'].mean() * 100  # 吃鸡率(%)

plt.figure(figsize=(10, 6))
win_by_level.plot(kind='bar', color='skyblue')
plt.title('大师局：不同等级的吃鸡率（Top1 %）')
plt.xlabel('玩家等级')
plt.ylabel('吃鸡率 (%)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
